```python
import pandas as pd
```

```python
data = pd.read_csv('dataset/training_data.csv')
```

```python
# Display basic information about the dataset
print("Shape of the dataset:", data.shape)
print("\nFirst few rows of the dataset:\n", data.head())
```

```python
# Display the column names in your DataFrame
print(data.columns)
```

```python
data = data.dropna(axis='columns')
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming 'data' is your DataFrame and 'categorical_column' is the column to be encoded
label_encoder = LabelEncoder()
data['prognosis'] = label_encoder.fit_transform(data['prognosis'])
```

```python
# Assuming 'categorical_column' is the column name containing categorical data
encoded_data = pd.get_dummies(data, columns=['prognosis'])
```

```python
# Assuming 'data' is your modified dataset after encoding categorical columns
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
# Combine imputed numeric data with the non-numeric 'prognosis' column
imputed_data = pd.DataFrame(data, columns=data.columns[:-1])  # Exclude the last column 'Unnamed: 133'
imputed_data['prognosis'] = data['prognosis']
```

```python
from sklearn.decomposition import PCA

# Separate the 'prognosis' column from the rest of the data
X = imputed_data.drop(columns=['prognosis'])

# Perform PCA
pca = PCA(n_components=2)
transformed_data = pca.fit_transform(X)
```

```python
import matplotlib.pyplot as plt

# Assuming transformed_data is the result of PCA transformation
plt.scatter(transformed_data[:, 0], transformed_data[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Transformed Data')
plt.show()
```

```python
# Assuming your data contains the symptom columns
symptoms = data.drop(columns=['prognosis']).columns.tolist()
```

```python
# Assuming pca is your fitted PCA model and data is your transformed data
# pca.fit(data)
# transformed_data = pca.transform(data)

# Get the principal components
components = pca.components_

# Assuming 'symptoms' contains the symptom names
# For example: symptoms = data.drop(columns=['prognosis']).columns

# Analyze the first principal component
component_num = 0  # Change this to explore different components
component = components[component_num]

# Pair each symptom name with its loading in the component
symptoms_loading = dict(zip(symptoms, component))

# Sort symptoms by their absolute loading value
sorted_symptoms = sorted(symptoms_loading.items(), key=lambda x: abs(x[1]), reverse=True)

# Display the top symptoms contributing to the principal component
top_n = 40  # Change this to display more or fewer symptoms
top_symptoms = sorted_symptoms[:top_n]

for symptom, loading in top_symptoms:
    print(f"{symptom}: {loading}")
```

```python
#assuming you've defined symptoms as a list of column names in your dataset
# components is the PCA components array obtained from the previous PCA fit

# Select a principal component number to analyze
component_num = 0  # Change this number based on your analysis

# Fetch the specified principal component
component = components[component_num]

# Pair each symptom name with its loading in the component
symptoms_loading = dict(zip(symptoms, component))

# Sort symptoms by their absolute loading value
sorted_symptoms = sorted(symptoms_loading.items(), key=lambda x: abs(x[1]), reverse=True)

# Extract the first 40 symptoms
top_40_symptoms = sorted_symptoms[:40]

# Print or examine the top 40 symptoms
print(top_40_symptoms)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you've extracted the top symptoms using the previous code
# top_40_symptoms contains the top symptoms along with their loading values

# Create a DataFrame from the top symptoms
import pandas as pd
df = pd.DataFrame(top_40_symptoms, columns=['Symptom', 'Loading Value'])

# Plotting the loading values of the top symptoms using Seaborn
plt.figure(figsize=(10, 8))
sns.barplot(x='Loading Value', y='Symptom', data=df, palette='viridis')
plt.xlabel('Loading Value')
plt.ylabel('Symptoms')
plt.title('Importance of Top Symptoms')
plt.show()
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Assuming you've extracted the top symptoms using the previous code
# top_40_symptoms contains the top symptoms along with their loading values

# Extract symptom names and their loading values separately
symptom_names = [symptom[0] for symptom in top_40_symptoms]
loading_values = np.array([symptom[1] for symptom in top_40_symptoms]).reshape(-1, 1)

# Normalizing the loading values
scaler = MinMaxScaler()
loading_values_normalized = scaler.fit_transform(loading_values).flatten()

# Plotting the loading values of the top symptoms
plt.figure(figsize=(12, 8))
plt.barh(symptom_names, loading_values_normalized, color='skyblue')
plt.xlabel('Normalized Loading Value')
plt.title('Importance of Top Symptoms (Normalized)')
plt.show()
```
